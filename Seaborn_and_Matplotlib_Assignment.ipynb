```python
#write code here
fare_bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]

fare_categories = pd.cut(titanic['fare'], bins=fare_bins, right=False)

fare_counts = fare_categories.value_counts().sort_index()

fare_counts_df = fare_counts.reset_index()
fare_counts_df.columns = ['Fare Range', 'Count']

plt.figure(figsize=(12, 6))
sns.barplot(x='Fare Range', y='Count', data=fare_counts_df, palette='viridis')

plt.title('Count of Passengers by Fare Range')
plt.xlabel('Fare Range')
plt.ylabel('Count of Passengers')

plt.show()

```

Question 12: Draw subplots containing 4 different graphs of your choice to show relation between different columns and set hue,title and Axes name for all graphs

```python
#write code here
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Scatter plot of Age vs Fare
sns.scatterplot(x='age', y='fare', hue='sex', data=titanic, ax=axes[0, 0])
axes[0, 0].set_title('Scatter plot of Age vs Fare')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Fare')

# Plot 2: Bar plot of Class vs Fare
sns.barplot(x='class', y='fare', hue='sex', data=titanic, ax=axes[0, 1])
axes[0, 1].set_title('Bar plot of Class vs Fare')
axes[0, 1].set_xlabel('Class')
axes[0, 1].set_ylabel('Fare')

# Plot 3: Box plot of Age by Class
sns.boxplot(x='class', y='age', hue='sex', data=titanic, ax=axes[1, 0])
axes[1, 0].set_title('Box plot of Age by Class')
axes[1, 0].set_xlabel('Class')
axes[1, 0].set_ylabel('Age')

# Plot 4: Count plot of Survived by Sex
sns.countplot(x='sex', hue='survived', data=titanic, ax=axes[1, 1])
axes[1, 1].set_title('Count plot of Survival by Sex')
axes[1, 1].set_xlabel('Sex')
axes[1, 1].set_ylabel('Count')


plt.show()
```

Question 13

Visualize the Survival Rate by Gender:

Using Matplotlib, create a bar plot to show the survival rate by gender.

Using Seaborn, create a count plot to show the survival count by gender and use the hue parameter to distinguish between survived and not survived passengers.


```python
#write code here
```

```python
survival_rate_gender = titanic.groupby('sex')['survived'].mean()

plt.figure(figsize=(8, 6))
survival_rate_gender.plot(kind='bar', color=['blue', 'pink'], alpha=0.7)

plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)

plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set1')

plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```
